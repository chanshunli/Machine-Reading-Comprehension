```python
from nltk import word_tokenize
import json
from tqdm import tqdm
import pandas as pd
import numpy as np
import pymongo
from pprint import pprint
from pymongo import MongoClient
from bson.objectid import ObjectId
from nltk.corpus import wordnet as wn
import requests
client = MongoClient('localhost', 27017)
db = client['fn_en_170']
```

```python
squad = pd.read_csv('./squad_da_question.csv')
squad.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
total = max(squad['num'].values.tolist())
print('>>> Nums of question:',total)
```

```python
data = {}
for num in tqdm(range(total)):
    try:
        one = squad[squad['num'] == num]
        sentence_length = max(one['word_index'].values.tolist())
        sample = one.head(sentence_length)
        words = sample['word'].values.tolist()

        s = ''
        for word in words:
            s += word + ' '
        s = s[:-3]+'?'
#         print('>>> %s'%s)
        
        feInfo = one[one['frame'] != '_']
        lus = feInfo['lu'].values.tolist()
        frames = feInfo['frame'].values.tolist()
        words = feInfo['word'].values.tolist()
        
        mydict = {s:[]}
        
        updatedict={}
        for lu_unit,frame_unit,word_unit in zip(lus,frames,words):
#             print('>>> word  : %s'%word_unit)
#             print('>>> lu    : %s'%lu_unit)
#             print('>>> frame : %s'%frame_unit)
            lexemeSet = []
            
            for query in db['frames'].find({'name':frame_unit}):
                lu_id=query['lexUnits']
            for luid in lu_id:
                for query in db['lexunits'].find({'_id':luid}):
                    pos = query['pos']
                    name = query['name']
                    status = query['status']
                    for query in db['lexunits'].find({'name':name}):
                        lexemes = query['lexemes'][0]
                    for query in db['lexemes'].find({'_id':lexemes}):
                        lexeme = query['name']
                        lexemeSet.append(lexeme)
            result = [s.replace(word_unit,w) for w in lexemeSet[:5]]
            updatedict[word_unit]=result
#             print(result) 
        
        data[s]=updatedict
        

        
#         print('-----------------------------------------------------------------')
    except:
        pass

with open('example.json','w') as f:
    json.dump(data,f)
    

    
    
```
