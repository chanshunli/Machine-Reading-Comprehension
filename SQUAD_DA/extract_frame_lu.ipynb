```python
import json
from tqdm import tqdm
import pandas as pd
import numpy as np
import pymongo
from pprint import pprint
from pymongo import MongoClient
from bson.objectid import ObjectId
client = MongoClient('localhost', 27017)
db = client['fn_en_170']
```

```python
'''

以lu:   son.n为例

_id:2144指向son.n
name:      son.n
pos :      N
lemmaid:指向son.n
frame:     104指向其他


---lexunits
       |------_id      
       |------name
       |------pos
       |------lemmaID
       |------frame
       |------status
       |------lexemes
       |------semTypes

'''
```

```python
print('>>> Who was the son of chiron ?')
for query in db['lexunits'].find({'name':'son.n'}):
    print('>>> LexUnits [%s]'%query['name'])
    lexemes = query['lexemes'][0]
    frame_id = query['frame']
for query in db['lexemes'].find({'_id':lexemes}):
    print('>>> Lexeme   [%s]'%query['name'])
for query in db['frames'].find({'_id':frame_id}):
    print('>>> frame    [%s]'%query['name'])
    fe_id=query['frameElements']
    lu_id=query['lexUnits']
print('----------------------------FrameElements-------------------------')
print('%5s      %22s    %7s    %15s'%('_id','name','requires','excludes'))
for feid in fe_id:
    for query in db['frameelements'].find({'_id':feid}):
        requires = query['requires']
        excludes = query['excludes']
        name = query['name']
        print('[%5s]    %22s    %7s    %15s'%(feid,name,requires,excludes))
print('------------------------------LexUnits----------------------------')  
print('%5s        %22s      %7s   %15s'%('_id','name','pos','status'))
for luid in lu_id:
    for query in db['lexunits'].find({'_id':luid}):
        pos = query['pos']
        name = query['name']
        status = query['status']
        print('[%5s]      %22s    %7s    %15s'%(luid,name,pos,status))
```
